```python
df = pd.read_csv('../../lvr/a_lvr_land_a.csv')
df_2 = pd.read_csv('../../lvr/a_lvr_land_a_build.csv')
df_3 = pd.concat([df, df_2])
df_3.head(3)
```

```python
#df_3 = df_3.drop(index = 0)
df_3.drop(columns=['建物分層','總層數'])
df_4 = df_3.dropna(subset=['交易標的', '交易年月日','單價元平方公尺'])  #留下可計算內容
df_4
```

```python
#df = df.drop(index = 0)
df = df[~df['交易標的'].isin(['車位'])] #排除單售車位
df = df[~df['都市土地使用分區'].isin(['商'])] 
df['車位類別'] = df['車位類別'].fillna('無')
df.head(2)
dummy_1 = pd.get_dummies(df['交易標的'])
dummy_2 = pd.get_dummies(df['移轉層次'])
dummy_3 = pd.get_dummies(df['有無管理組織'])
dummy_4 = pd.get_dummies(df['車位類別'])
dummy_5 = pd.get_dummies(df['鄉鎮市區'])
dummy_6 = pd.get_dummies(df['建物型態'])


df_total = df['總價元']
df_concat = pd.concat([df['土地移轉總面積平方公尺'],dummy_1,dummy_2,dummy_3,dummy_4,dummy_5,dummy_6],axis=1)
df_concat
```

```python
# build model 
from sklearn.linear_model import LinearRegression
from sklearn import datasets
from sklearn.model_selection import train_test_split
X = df_concat
y = df_total 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
reg = LinearRegression().fit(X_train, y_train)
print(reg.score(X_train, y_train)) #R^2 值
print(reg.intercept_) #截距
print(reg.coef_)#自變數係數
```

```python
#regresion predict 

predictions = reg.predict(X_test)

print(y_test) # the real number 
print(predictions[0]) # the predition number 
#df_predict = pd.DataFrame(predictions)
```

```python
###### cluster #####
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist #use distance as the as the measured index 
distortions = []
K = range(1,15)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X_train)
    kmeanModel.fit(X_train)
    distortions.append(sum(np.min(cdist(X_train, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X_train.shape[0])

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
# 8 class is better 
k_model = KMeans(n_clusters=8).fit(X_train)
k_model.labels_ #cluster groop number 
```
